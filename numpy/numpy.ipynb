#####Be careful when copying arrays!!

#####When you do b = a, both variables point to the same array in memory. Changing b also changes a. Use b = a.copy() to create an independent copy.

```python
a = np.array([1,2,3])
b = a
b
b[0] = 22
print(a)
```

```python
b = a.copy()
b
```

##Reorganizing arrays

```python
before = np.array([[1,2,3,4], [5,6,7,8]])
print(before)

#The product of the new shape dimensions must equal the total number of elements in the original array.
after = before.reshape((4,2))
print(after)
```

```python
#Vertically Stacking Vectors

v1 = np.array([1,2,3,4])
v2 = np.array([5,6,7,8])

np.vstack([v1,v2,v1,v2])
```

```python
#Horizontal Stack
```

```python
h1 = np.ones((2,4))
h2 = np.zeros((2,3))
print(h1)
print(h2)

np.hstack([h1,h2])
```
